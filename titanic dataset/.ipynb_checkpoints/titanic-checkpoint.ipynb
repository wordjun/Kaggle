```python
import matplotlib.pyplot as plt
#주피터 노트북 안에 차트가 그려지도록 하기
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
plt.style.use('seaborn')
sns.set(font_scale=2.5)
# matplotlib의 기본 scheme 말고 seaborn scheme를 세팅하고, 
# 일일이 graph 의 font size를 지정할 필요 없이 seaborn의 font_scale을 사용하면 편하다.
import missingno as msno # conda install -c conda-forge missingno로 conda prompt에서 설치해야 합니다.
#ignore warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')
```

```python
train_data.head()
```

Features:
-Pclass(ticket class)
-Age
-silbSp(num. of siblings and spouse)
-parch(no.of parents & children on board together)
-fare
-target label to predict: Survived(0 or 1, one-hot encoded)

```python
train_data.describe()
```

데이터 카운트는 891개인데 Age 속성의 개수는 714이므로 null값인 데이터가 존재함을 알 수 있다

```python
test_data.describe()
```

test데이터도 마찬가지(passengerid는 418개, age는 332개)

```python
print(train_data.columns) #train데이터의 column, 즉 모든 feature들의 이름 출력
```

```python
for col in train_data.columns:
    msg = "column: {:>10}\t Percent of NaN value: {:2f}%".format(
    col, 100  * (train_data[col].isnull().sum() / train_data[col].shape[0]))
    print(msg)
```

train 데이터셋에서 age의 약 20%정도, cabin의 77%정도가 null값임을 알 수 있다

```python
for col in test_data.columns:
    msg = "column: {:>10}\t Percent of NaN value: {:2f}%".format(
    col, 100  * (test_data[col].isnull().sum() / test_data[col].shape[0]))
    print(msg)
```

test 데이터셋 역시 age의 약 20%정도, cabin의 78%정도가 null값임을 알 수 있다

```python
#msno.matrix를 활용해 행렬 형태로 데이터를 시각화함
#figsize는 (가로, 세로)
#color는 각각 RGB값
msno.matrix(df = train_data.iloc[:, :], figsize=(8, 8), color=(0.8, 0.1, 0.5))
```

```python
msno.bar(df = train_data.iloc[:, :], figsize=(8, 8), color=(0.3, 0.1, 0.5)) #bar그래프 형태로 더 쉽게 알 수 있다.
```

```python
#target label인 Survived가 어떤 분포도를 이루는지 확인해야 하므로
#지금같은 binary classification문제의 경우에서 0과 1의 분포가 어떠냐에 따라
#모델의 평가방법이 달라질 수 있다.
f, ax = plt.subplots(1, 2, figsize=(18, 8)) #figsize로 크기지정
train_data['Survived'].value_counts().plot.pie(explode=[0, 0.1], autopct='%1.1f%%', ax = ax[0], shadow=True)
ax[0].set_title('Pie plot - survived')
ax[0].set_ylabel('')
sns.countplot('Survived', data=train_data, ax=ax[1])
ax[1].set_title('Count plot - Survived')
```

```python
#Pclass는 ordinal(서수형)데이터이다; category이면서도 순서가 있음
#Pclass에 따른 생존률 차이
#count()로 각 class에 몇명이 있는지 확인
train_data[['Pclass', 'Survived']].groupby(['Pclass']).count()
#합계891명 확인
```

```python
#sum()으로 총 몇명이 생존했는지 확인
train_data[['Pclass', 'Survived']].groupby(['Pclass']).sum()
#생존자 총 891명 중  342명, 생존률 약 38%
```

```python

pd.crosstab(train_data['Survived'], train_data['Pclass'], margins=True).style.background_gradient(cmap='summer_r')
```

```python
train_data[['Pclass', 'Survived']].groupby(['Pclass'], as_index=True).mean().sort_values(by='Survived', ascending=True).plot.bar()
#pclass가 좋을수록 생존율이 높다. 당시 상황은 1등급칸의 부유한 사람들 먼저 대피시켰고 3등급의 일반 시민들은 나중에 태웠기 때문에
#결국 3등급의 생존율이 월등히 낮은 것을 확인할 수 있다.
```

```python
#plt.subplots(1, 2, figsize=(18, 8)) #1행 2열로 두개의 그래프 생성. 크기는 (18 x 8)
# 타이틀이 그래프 위에 얼마나 떨어졌는지 정하는 상수 값
f, ax = plt.subplots(1, 2, figsize=(18, 8))
#ax[0]은 왼쪽 그래프를 뜻하고, ax[1]은 오른쪽 그래프를 뜻합니다.
train_data['Pclass'].value_counts().plot.bar(color=[0, 0, 0], ax=ax[0])
# ax[0]이므로 왼쪽에 그래프를 그립니다.
ax[0].set_title('Number of Passengers By Pclass', y=1)
ax[0].set_ylabel('Count')
sns.countplot('Pclass', hue='Survived', data=train_data, ax=ax[1])
# ax[1]이므로 오른쪽에 그래프를 그립니다.
ax[1].set_title('Pclass: Survived vs Dead', y=1)
plt.show()

#클래스가 높을수록 생존확률이 높고, 생존률에 pclass가 큰 영향을 미친다고 생각할 수 있음.
```

```python
f, ax = plt.subplots(1, 2, figsize=(18, 8))
#ax[0]은 왼쪽 그래프를 뜻하고, ax[1]은 오른쪽 그래프를 뜻합니다.
train_data[['Sex', 'Survived']].groupby(['Sex']).mean().plot.bar(ax=ax[0])
ax[0].set_title('Survived vs. Sex')
sns.countplot('Sex', hue='Survived', data=train_data, ax=ax[1])

#female had higher chance of survival
```

```python
train_data[['Sex', 'Survived']].groupby(['Sex']).mean().sort_values(by = 'Survived', ascending=False)
```

```python
#margins shows the total amount of each attributes
pd.crosstab(train_data['Sex'], train_data['Survived'], margins=True).style.background_gradient(cmap='summer_r')
```

```python
sns.factorplot('Pclass', 'Survived', hue='Sex', data=train_data, size=6, aspect=2)
#모든 class에서 여자가 살 확률이 높음을 알 수 있다.
```

```python
sns.factorplot('Sex', 'Survived', col='Pclass', data=train_data, size=6, aspect=1)
```

```python
print("Oldest passenger : {:.1f} years old".format(train_data['Age'].max()))
print("Youngest passenger : {:.1f} years old".format(train_data['Age'].min()))
print("Average age of passengers : {:.1f} years old".format(train_data['Age'].mean()))
```

```python
f, ax = plt.subplots(1, 1, figsize=(9, 5))
sns.kdeplot(train_data[train_data['Survived']==1]['Age'], ax =ax)
sns.kdeplot(train_data[train_data['Survived']==0]['Age'], ax =ax)
plt.legend(['Survived', 'Dead'])
```

```python
sns.relplot(data = train_data, x = 'Age', y = 'Survived', hue = 'Sex', aspect=3)
```

```python
sns.catplot(data=train_data, x='Pclass', y = 'Fare', aspect = 4)
```
